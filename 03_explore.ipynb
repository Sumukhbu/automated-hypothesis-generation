```python
# 03_explore.ipynb

import os
import json
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load cleaned abstracts
with open("data/processed/cleaned_abstracts.json", "r", encoding="utf-8") as f:
    papers = json.load(f)

# Flatten all tokens
all_tokens = [token for paper in papers for token in paper["tokens"]]

# -------------------------------
# 1. Top 20 most common words
# -------------------------------
counter = Counter(all_tokens)
top_words = counter.most_common(20)

print("ðŸ”Ž Top 20 words:")
for word, count in top_words:
    print(f"{word}: {count}")

# Plot bar chart
plt.figure(figsize=(10, 5))
plt.bar([w for w, _ in top_words], [c for _, c in top_words])
plt.xticks(rotation=45)
plt.title("Top 20 Most Common Words")
plt.show()

# -------------------------------
# 2. Word cloud
# -------------------------------
wc = WordCloud(width=800, height=400, background_color="white").generate(" ".join(all_tokens))

plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Abstracts")
plt.show()

# -------------------------------
# 3. Distribution of abstract lengths
# -------------------------------
lengths = [len(paper["tokens"]) for paper in papers]

plt.figure(figsize=(8, 5))
plt.hist(lengths, bins=10, color="skyblue", edgecolor="black")
plt.title("Distribution of Abstract Lengths (in tokens)")
plt.xlabel("Number of tokens")
plt.ylabel("Frequency")
plt.show()

print(f"ðŸ“Š Average abstract length: {sum(lengths)/len(lengths):.2f} tokens")
```
